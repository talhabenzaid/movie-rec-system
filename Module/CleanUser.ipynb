```python
import pandas as pd

UserDataset = pd.read_csv('../Data/user/u.data', 
                      sep='\t', names=['user_id', 'movie_id', 'rating', 'timestamp'], engine='python')
UserDataset.head()
```

```python
UserDataset = UserDataset.drop('timestamp', axis=1)
UserDataset.isna().sum()
```

```python
duplicates = UserDataset[UserDataset.duplicated(subset=['user_id', 'movie_id'], keep=False)]
print("Number of duplicate ratings:", len(duplicates))
```

```python
UserDataset2 = pd.read_csv(
    "../Data/user/u.item",
    sep='|',
    encoding='latin-1',
    usecols=[0, 1],
    names=['movie_id', 'title'],
    header=None
)

def remove_year(title):
    result = ""
    for s in title:
        if s == "(":
            break
        result += s
    return result.strip()

UserDataset2['title'] = UserDataset2['title'].apply(remove_year)

UserDataset2.head()
```

```python

NewUser = UserDataset.merge(UserDataset2, on='movie_id', how='left')
NewUser.head()
```

```python
NewUser.to_csv("../NewData/NewUser.csv",index=False)
```
