```python
import pandas as pd
DataKey  = pd.read_csv("../Data/keyword/tmdb_5000_movies.csv")
columns_to_keep = ['id', 'title', 'genres', 'overview', 'keywords', 'vote_average', 'release_date']

DataKey = DataKey[columns_to_keep]

DataKey.head()
```

```python
DataKey = DataKey.dropna()
print(DataKey.isna().sum())
```

```python
import json

def parse_json_column(column_value):
    try:
        names = []
        data = json.loads(column_value)
        for item in data:
            if 'name' in item:
                names.append(item['name'])
        return ' '.join(names)
    except (json.JSONDecodeError, TypeError):
            return ''
        
for column in ['genres', 'keywords']:
    DataKey[column] = DataKey[column].apply(parse_json_column)
DataKey['year'] = pd.to_datetime(DataKey['release_date'], errors='coerce').dt.year
DataKey = DataKey.drop('release_date', axis=1)

DataKey.head()
```

```python
DataKey.to_csv('../NewData/NewKeyword.csv', index=False)
```
